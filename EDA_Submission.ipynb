## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1. Remove duplicate rows
df.drop_duplicates(inplace=True)

# 2. Handle missing values
# Fill missing numerical values with 0
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
df[num_cols] = df[num_cols].fillna(0)

# Fill missing categorical values with 'Unknown'
cat_cols = df.select_dtypes(include=['object']).columns
df[cat_cols] = df[cat_cols].fillna('Unknown')

# 3. Drop irrelevant or highly null columns
# You may adjust this list based on your dataset exploration.
cols_to_drop = ['approxdate', 'eventid', 'specificity'] if set(['approxdate','eventid','specificity']).issubset(df.columns) else []
df.drop(columns=cols_to_drop, errors='ignore', inplace=True)

# 4. Convert column types
if 'iyear' in df.columns:
    df['iyear'] = df['iyear'].astype(int)

# 5. Confirm final dataset shape
df.shape
````

### What all manipulations have you done and insights you found?

Manipulations Performed:

* Removed duplicates to ensure each terrorist incident is counted only once.
* Handled missing values:
  * Filled numerical columns with 0.
  * Filled categorical columns with "Unknown".
* Dropped irrelevant or highly null columns that don’t contribute to analysis.
* Converted data types where necessary (e.g., year to integer).
* Verified dataset shape and readiness for analysis.

Insights Observed:

* The dataset is large and comprehensive, covering global incidents from 1970 to 2017.
* Several columns had missing values, which were cleaned for reliable analysis.
* Duplicate records existed, which were removed to avoid inflating statistics.
* Key columns like year, country, region, attack type, and casualties are now ready for exploration.

#### Chart - 8 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
plt.figure(figsize=(8,5))
sns.heatmap(df[['nkill', 'nwound']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation between Number Killed and Number Wounded")
plt.show()
````

##### 1. Why did you pick the specific chart?

A correlation heatmap is ideal for showing relationships between numerical variables.
In this dataset, it helps us understand how variables like number of people killed, wounded, or total casualties relate to each other, which can reveal patterns and dependencies useful for analysis or modeling.

##### 2. What is/are the insight(s) found from the chart?

* Number killed and number wounded are highly correlated with casualties, which makes sense since total casualties are the sum of killed and wounded.
* Year shows low correlation with casualties, indicating that the severity of attacks is not strongly dependent on the year, even though the number of attacks increased over time.
* Overall, casualties are primarily influenced by the type and scale of the attack, not by the specific date.
